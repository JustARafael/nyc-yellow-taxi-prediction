Getting the day of week for each row

We will be basing it on the pickup_datetime and adding it as a new column called weekday

0 = Monday

1 = Tuesday 

   ...
   
6 = Sunday

```python
january_timestamp = pd.to_datetime(pd.Series(january2013['pickup_datetime']))
january_timestamp_array = []
for index in january_timestamp:
    january_timestamp_array.append(index.dayofweek)
january2013['weekday'] = january_timestamp_array
print("Finished January")

"""
february_timestamp = pd.to_datetime(pd.Series(february2013['pickup_datetime']))
february_timestamp_array = []
for index in february_timestamp:
    february_timestamp_array.append(index.dayofweek)
february2013['weekday'] = february_timestamp_array
print("Finished February")

march_timestamp = pd.to_datetime(pd.Series(march2013['pickup_datetime']))
march_timestamp_array = []
for index in march_timestamp:
    march_timestamp_array.append(index.dayofweek)
march2013['weekday'] = march_timestamp_array
print("Finished March")

april_timestamp = pd.to_datetime(pd.Series(april2013['pickup_datetime']))
april_timestamp_array = []
for index in apriltimestamp:
    april_timestamp_array.append(index.dayofweek)
april2013['weekday'] = april_timestamp_array
print("Finished April")
"""
```

Subset by day of the week and verify the shapes to make sure its equal

```python
january2013monday = pd.DataFrame(january2013)
january2013monday = january2013monday.loc[january2013monday['weekday'] == 0]
january2013monday.reset_index()

january2013tuesday = pd.DataFrame(january2013)
january2013tuesday = january2013tuesday.loc[january2013tuesday['weekday'] == 1]
january2013tuesday.reset_index()

january2013wednesday = pd.DataFrame(january2013)
january2013wednesday = january2013wednesday.loc[january2013wednesday['weekday'] == 2]
january2013wednesday.reset_index()

january2013thursday = pd.DataFrame(january2013)
january2013thursday = january2013thursday.loc[january2013thursday['weekday'] == 3]
january2013thursday.reset_index()

january2013friday = pd.DataFrame(january2013)
january2013friday = january2013friday.loc[january2013friday['weekday'] == 4]
january2013friday.reset_index()

january2013saturday = pd.DataFrame(january2013)
january2013saturday = january2013saturday.loc[january2013saturday['weekday'] == 5]
january2013saturday.reset_index()

january2013sunday = pd.DataFrame(january2013)
january2013sunday = january2013sunday.loc[january2013sunday['weekday'] == 6]
january2013sunday.reset_index()
```

```python
print("Original shape is ", january2013.shape)
print("Monday's shape is ", january2013monday.shape)
print("Tuesday's shape is ", january2013tuesday.shape)
print("Wednesday's shape is ", january2013wednesday.shape)
print("Thursday's shape is ", january2013thursday.shape)
print("Friday's shape is ", january2013friday.shape)
print("Saturday's shape is ", january2013saturday.shape)
print("Sunday's shape is ", january2013sunday.shape)
```
